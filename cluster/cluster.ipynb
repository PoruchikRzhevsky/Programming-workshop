```python
#Importing the libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Loading the data

data = pd.read_csv('Terzan 2 VI photometry (Ortolani+ 1997).csv')
```

```python
data
```

```python
# Plot the data and the best-fit line

plt.rc("font", size=10)
plt.rcParams["font.family"] = "Times New Roman"
fig, ax = plt.subplots(figsize=(10, 10))

# ax.title.set_fontsize(15)
ax.set_xlabel(r"V-I [mag]")
ax.xaxis.label.set_fontsize(20)
ax.set_ylabel(r"V [mag]")
ax.yaxis.label.set_fontsize(20)
ax.invert_yaxis()
ax.tick_params(axis="both", which="major", length=10, width=0.5, labelsize=15)
ax.tick_params(axis="both", which="minor", length=5, width=0.5, labelsize=15)
# plt.xticks(rotation=45)
plt.tight_layout()  

ax.scatter(data['V-I'], data['Vmag'], color='blue')

plt.show() 

# fig.savefig('CMD.png', bbox_inches='tight')
```
