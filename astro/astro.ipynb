```python
#Importing the libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Loading the data

data = pd.read_csv(r'ZA2_t06.txt', sep=r'\s+')
```

```python
data
```

```python
# Plot the data and the best-fit line

plt.rc("font", size=10)
plt.rcParams["font.family"] = "Times New Roman"
fig, ax = plt.subplots(figsize=(10, 10))

# ax.title.set_fontsize(15)
ax.set_xlabel(r"log(P) [days]")
ax.xaxis.label.set_fontsize(20)
ax.set_ylabel(r"m [mag]")
ax.yaxis.label.set_fontsize(20)
ax.invert_yaxis()
ax.tick_params(axis="both", which="major", length=10, width=0.5, labelsize=15)
ax.tick_params(axis="both", which="minor", length=5, width=0.5, labelsize=15)
plt.tight_layout()  

ax.scatter(data['logP'], data['M'], color='blue')

plt.show() 

fig.savefig('cefeidy.png', bbox_inches='tight')
```
